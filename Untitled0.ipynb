```python
# necessary libraries
import pinecone
import pandas as pd
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Pinecone
from langchain.chains import RetrievalQA
from langchain.llms import HuggingFacePipeline
from transformers import pipeline
import torch
```

```python
from pinecone import Pinecone, ServerlessSpec

pc = Pinecone(
    api_key="PINECONE_API_KEY",
    environment="PINECONE_API_ENVI"
)

index_name = "quickstart"

try:
    if index_name not in pc.list_indexes().names():
        pc.create_index(
            name=index_name,
            dimension=768,
            metric="cosine",
            spec=ServerlessSpec(
                cloud="aws",
                region="PINECONE_API_ENVI"
            )
        )
        print(f"Index {index_name} created successfully.")
    else:
        print(f"Index {index_name} already exists.")

except Exception as e:
    print(f"An error occurred: {e}")

index = pc.Index(index_name)
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Amazon Sale Report.csv")
```

```python
# Text transformation function
def row_to_text(row):

    color_info = f", Color: {row['Color']}" if 'Color' in row else ""
    currency_info = f" {row['Currency']}" if 'Currency' in row else ""
    return (
        f"On {row['Date']}, {row['Qty']} units of {row['Category']} (Size: {row['Size']}{color_info}) "
        f"were sold via {row['Fulfilment']} for {row['Amount']}{currency_info}."
    )

# transformation to create text summaries
data['text'] = data.apply(row_to_text, axis=1)
documents = data[['text']].to_dict(orient="records")
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from transformers import AutoModel, AutoTokenizer
from concurrent.futures import ThreadPoolExecutor
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

model_name = "sentence-transformers/all-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name).to(device)

def embed_text(text):
    tokens = tokenizer(text, truncation=True, padding=True, return_tensors="pt").to(device)
    embeddings = model(**tokens).last_hidden_state.mean(dim=1)
    return embeddings.cpu().detach().numpy().flatten()

def preprocess_text(doc_text):
    return doc_text[:512]

documents = [{"text": preprocess_text(doc["text"])} for doc in documents]

batch_size = 100

def process_batch(start_idx, batch_docs):
    batch_upserts = [
        (str(start_idx + idx), embed_text(doc["text"]), {"text": doc["text"]})
        for idx, doc in enumerate(batch_docs)
    ]
    index.upsert(batch_upserts)
\
def process_documents_in_parallel(documents):
    with ThreadPoolExecutor(max_workers=8) as executor:
        for i in range(0, len(documents), batch_size):
            batch_docs = documents[i:i + batch_size]
            executor.submit(process_batch, i, batch_docs)

process_documents_in_parallel(documents)
```

```python
from huggingface_hub import login

# Log in to Hugging Face with your token
login(token="YOUR_huggingface_API")

from transformers import pipeline
from langchain.chains import RetrievalQA
from langchain.vectorstores import Pinecone
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.llms import HuggingFaceHub
import os
os.environ["WANDB_DISABLED"] = "true"

qa_model = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/msmarco-distilbert-base-v3")

os.environ["PINECONE_API_KEY"] = "PINECONE_API_KEY"
os.environ["PINECONE_ENVIRONMENT"] = "PINECONE_ENVIRONMENT"

retriever = Pinecone.from_existing_index(index_name="quickstart", embedding=embedding_model)

qa_chain = RetrievalQA.from_chain_type(
    llm=HuggingFaceHub(
        repo_id="google/flan-t5-small",
        huggingfacehub_api_token="YOUR_huggingface_API",
    ),
    retriever=retriever.as_retriever(),
    return_source_documents=True,
)

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

#queries and labels for classification (1: requires retrieval, 0: direct answer)
queries = ["How many units of electronics were sold in December?", "What is the highest selling category?"]
labels = [1, 0]

# Vectorize queries for classification
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(queries)
classifier = LogisticRegression()
classifier.fit(X, labels)

# Classify the incoming query
query = "How many units of electronics were sold?"
X_query = vectorizer.transform([query])
is_retrieval = classifier.predict(X_query)[0]

# Handle the query based on classification
if is_retrieval == 1:
    result = qa_chain.invoke({"query": query})
    print("Retrieved answer:", result["result"])
    print("Source documents:", result["source_documents"])
else:
    context = "Direct answer: 50 units of electronics."
    answer = qa_model(question=query, context=context)
    print("Direct answer:", answer["answer"])

# Fine-tuning a T5 model for your use case
from datasets import Dataset
from transformers import Trainer, TrainingArguments, AutoModelForSeq2SeqLM, AutoTokenizer

# Example fine-tuning dataset
train_data = {
    "input": ["How many units of electronics were sold in December?"],
    "output": ["50 units of electronics were sold in December."]
}
dataset = Dataset.from_dict(train_data)


# Load T5 model and tokenizer
model_name = "t5-small"
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Tokenize the dataset
def preprocess(examples):
    inputs = examples['input']
    outputs = examples['output']
    model_inputs = tokenizer(inputs, text_target=outputs, truncation=True, padding="max_length", max_length=512)
    return model_inputs

tokenized_dataset = dataset.map(preprocess, batched=True)

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="no",
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    num_train_epochs=3,
    weight_decay=0.01,
    save_steps=10,
    save_total_limit=2,
    report_to="none"
)

# Initialize Trainer and train the model
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
)

trainer.train()
```
